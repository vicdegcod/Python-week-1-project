```python
import datetime

def get_birth_year():
    try:
        age = input("How old are you? ")
        age_int = int(age) # Potential ValueError
        
        current_year = 2024
        birth_year = current_year - age_int
        
        print(f"You were born around {birth_year}")
        
    except ValueError:
        print("Error: You didn't enter a valid number for your age.")

# If we don't use try/except, the program dies here if input is "Ten"
get_birth_year() 
print("Program continues to run other tasks...")
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    
    a, b = 0, 1

    for i in range(2, n + 1):
        a, b = b, a + b

    return b

print(fibonacci(4))
```
